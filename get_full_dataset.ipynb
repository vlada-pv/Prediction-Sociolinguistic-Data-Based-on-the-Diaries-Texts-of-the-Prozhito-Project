**Full dataset**

```python
# Общий датасет, куда будем добавлять информацию
final_dataset = pd.DataFrame(columns=['diary', 'sex', 'birth'])
```

```python
unique_diary_indexes = notes['diary'].loc[notes['diary'].apply(lambda x: str(x).isnumeric())].unique()
```

```python
for diary_index in unique_diary_indexes:
    try:
        # Convert diary_index to integer
        diary_index = int(diary_index)

        # Получаем значение person из diaries.csv
        person_value = diaries.loc[diaries['id'] == diary_index, 'person'].values[0]

        # Получаем sex и birth из persons.csv if person_info is not empty
        person_info = persons.loc[persons['id'] == person_value, ['sex', 'birth']]

        final_dataset = pd.concat([final_dataset, pd.DataFrame({'diary': [diary_index], 'sex': [person_info['sex'].values[0]], 'birth': [person_info['birth'].values[0]]})], ignore_index=True)

    except IndexError:
        # Handle the IndexError here (e.g., set default values or skip this iteration)
        print(f"IndexError occurred for diary index {diary_index}. Skipping this entry.")
```

```python
final_dataset = final_dataset.rename(columns={'diary_num': 'diary'})
final_dataset
```

```python
notes = notes[pd.to_numeric(notes['diary'], errors='coerce').notna()]
notes['diary'] = notes['diary'].astype(int)
```

```python
for index, row in notes.iterrows():
    diary_value = row['diary']
    if isinstance(diary_value, (int, float)):
        matching_row = final_dataset[final_dataset['diary'] == diary_value]
        if not matching_row.empty:
            sex_value = matching_row['sex'].values[0]
            birth_value = matching_row['birth'].values[0]
            notes.loc[index, 'sex'] = sex_value
            notes.loc[index, 'birth'] = birth_value
    else:
        print(f"Значение в строке {index} столбца 'diary' не является числовым.")
```

```python
notes = notes.dropna()
notes
```

```python
from dateutil.parser import ParserError

# Создание пустого списка для хранения возрастов
ages = []

# Проход по каждой строке данных
for index, row in notes.iterrows():
    try:
        # Преобразование столбцов date и birth в формат даты
        date = pd.to_datetime(row['date'])
        birth = pd.to_datetime(row['birth'])

        # Вычисление возраста и добавление его в список ages
        age = date.year - birth.year
        ages.append(age)
    except (ValueError, ParserError):
        # Пропуск строки при ошибке преобразования даты
        ages.append(None)

# Добавление списка возрастов в датасет как новый столбец age
notes['age'] = ages
```
