## **Создание столбца с сохранением в токенах глаголов в прошедшем времени**

```python
! pip install pymorphy2
```

```python
import pymorphy2
from tqdm.auto import tqdm, trange
```

```python
# Создание анализатора
morph = pymorphy2.MorphAnalyzer()

# Функция для поиска глаголов прошедшего времени
def find_past_tense_verbs(text):
    words = text.split()  # Разделение текста на слова
    past_verbs = []
    for word in words:
        parsed_word = morph.parse(word)[0]  # Морфологический анализ слова
        if 'VERB' in parsed_word.tag and 'past' in parsed_word.tag:  # Проверка на глагол прошедшего времени
            past_verbs.append(word)
    return ' '.join(past_verbs)
```

```python
# Применение функции ко всему столбцу 'text' и соединение с 'text_preprocessing'
tqdm.pandas(desc="Processing rows")
data['text_preprocessing_past_verbs'] = data.progress_apply(lambda row: str(row['text_preprocessing']) + ' ' + find_past_tense_verbs(row['text']) if isinstance(row['text_preprocessing'], str) else find_past_tense_verbs(row['text']), axis=1)
```

```python
data.to_csv('/kaggle/working/data_final_with_past_verbs.csv',index=False)
```
